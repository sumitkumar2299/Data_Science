```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.datasets import make_blobs 
# mmake_blobs cluster generate kar ke deta hain 
from sklearn.cluster import KMeans 
from sklearn.preprocessing import StandardScaler 

```

```python
X,y_true = make_blobs(n_samples=500,centers=3,cluster_std=60,random_state=42)
# centre = 3 means you have three cluster 
```

```python
X
```

```python
df = pd.DataFrame(X,columns=['feature_1','feature_2'])
```

```python
df
```

```python
scaler = StandardScaler()
X_scale = scaler.fit_transform(df)
```

```python
# elbow method 

inertia = []
K_range = range(1,11)
```

```python
for k in K_range: 
    Kmeans = KMeans(n_clusters=k,random_state=42)
    Kmeans.fit(X_scale)
    inertia.append(Kmeans.inertia_)
```

```python
inertia 
```

```python
plt.plot(K_range,inertia,marker = 'o')
```

```python
Kmeans_final = KMeans(n_clusters=3,random_state=42)
```

```python
cluster_labels = Kmeans_final.fit_predict(X_scale)
```

```python
cluster_labels
```

```python
df['cluster'] = cluster_labels
```

```python
sns.scatterplot(x= df['feature_1'],y= df['feature_2'],hue =df['cluster'],palette='viridis')
```
