```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.datasets import make_blobs 
# mmake_blobs cluster generate kar ke deta hain 
from sklearn.cluster import KMeans 
from sklearn.preprocessing import StandardScaler 

```

```python
X,y_true = make_blobs(n_samples=500,centers=3,cluster_std=60,random_state=42)
# centre = 3 means you have three cluster 
```

```python
X
```

```python
df = pd.DataFrame(X,columns=['feature_1','feature_2'])
```

```python
df
```

```python
scaler = StandardScaler()
X_scale = scaler.fit_transform(df)
```

```python
# elbow method 

inertia = []
K_range = range(1,11)
```

```python
for k in K_range: 
    Kmeans = KMeans(n_clusters=k,random_state=42)
    Kmeans.fit(X_scale)
    inertia.append(Kmeans.inertia_)
```

```python
inertia 
```

```python
plt.plot(K_range,inertia,marker = 'o')
```

```python
Kmeans_final = KMeans(n_clusters=3,random_state=42)
```

```python
cluster_labels = Kmeans_final.fit_predict(X_scale)
```

```python
cluster_labels
```

```python
df['cluster'] = cluster_labels
```

```python
sns.scatterplot(x= df['feature_1'],y= df['feature_2'],hue =df['cluster'],palette='viridis')
```

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
```

```python
X, y_true = make_moons(n_samples=500,noise = 0.05,random_state=42)
```

```python
df = pd.DataFrame(X,columns = ['feature_1','feature_2'])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
kmeans = KMeans(n_clusters=2,random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)
```

```python
df['kmeans_cluster'] = kmeans_labels
```

```python
sns.scatterplot(x =df['feature_1'],y= df['feature_2'],hue = df['kmeans_cluster'],palette='tab10')
```

```python
dbscan = DBSCAN(eps = 0.3, min_samples = 5 )
dbscan_labels = dbscan.fit_predict(X_scaled)
```

```python
df['dbscan_cluster'] = dbscan_labels
```

```python
sns.scatterplot(x =df['feature_1'],y= df['feature_2'],hue = df['dbscan_cluster'],palette='tab10')
```
