```python
import numpy as np 
import seaborn as sns
from scipy.stats import chi2_contingency
import pandas as pd 
```

```python
df = sns.load_dataset("titanic")
```

```python
df 
```

```python
contingency_table = pd.crosstab(df['sex'], df['survived'])
contingency_table 
```

```python
chi2, p_value, dof , expected = chi2_contingency(contingency_table)
```

```python
expected
```

```python
chi2
```

```python
alpha = 0.05
```

```python
if p_value < alpha: 
    print("we reject the null hypothesis and there is a significant relationship between gender and survivor")
else: 
    print("there is no connection")
```
