```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot")
# you can choose any style that you want 
```

```python
# Basic pie charts examples 

# data 
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]

plt.title("career runs of indian batsman")
plt.pie(runs,labels = labels)
plt.show()
```

```python
# i can choose colors name or hex code by myself 

colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(runs,labels = labels, colors = colors)
plt.show()

```

```python
# Add edges and slices with wedgegroups 

plt.pie(
    runs,
    labels = labels,
    colors = colors,
    wedgeprops = {'edgecolor':'black','linewidth':2,'linestyle': '--'}
)
plt.show()
```

```python
# Highlights a slice using explode 

# only highlights sachin slice 
explode = [0.1,0,0,0]
plt.pie(runs,labels = labels, explode = explode, colors = colors)
plt.title("explode pie example")
plt.show()
```

```python
# add shadows for a 3d feel 

plt.pie(
    runs,
    labels = labels,
    explode = explode,
    colors = colors,
    # adds a 3d slice like shadow
    shadow = True  
)
plt.show()
```

```python
# startangele 
# use startangle to rotate the pie for better alignment 

plt.pie(
    runs,
    labels = labels,
    explode = explode,
    colors = colors,
    shadow = True,
    startangle = 140
)

plt.show()
```

```python
# show percentage with autopct 

plt.pie(
    runs,
    labels = labels,
    autopct = '%1.1f%%',
    startangle = 140
)
plt.title("carrer run share")
plt.show()
```

```python
# crowded pie charts 
# why to avoided 

# too many categories 
languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#', 'Ruby', 'Go', 'Rust',
'Swift', 'PHP']
usage = [30, 20, 10, 10, 7, 5, 4, 3, 2, 1]

plt.pie(usage, labels = languages, autopct = '%1.1f%%',startangle = 90)
plt.title ("programing language usage (crowded example)")
plt.show()

```

```python
# why avoid it?

# -> hard to compare slice sizes visually 
# -> cluttered and confusing
# -> no clear insight 
# -> better alternatives: bar charts or horizontal bar charts 
```

```python
# summary of useful pie charts parameters 



labels ->  Label each slice
colors -> Customize slice colors (names or hex)
explode ->  Pull out slices for emphasis
shadow ->  Adds depth-like shadow
startangle -> Rotates pie to start from a different angle
autopct ->  Shows percentage text on slices
wedgeprops ->  Customize slice edge, fill, width, etc.
```
